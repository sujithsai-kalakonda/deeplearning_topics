# a. Batch Vs Stocahstic

## 1. Fastness b/w Batch vs Stochastic

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(10, activation='relu', input_dim=2))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

#### BATCH GRADIENT DESCENT:

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
start = time.time()
history = model.fit(X_train, y_train, epochs=10, batch_size=320) # here batch_size=320 i.e, considering the entire data
print(time.time() - start)
```

#### STOCHSTIC GRADIENT DESCENT

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
start = time.time()
history = model.fit(X_train, y_train, epochs=10, batch_size=1) # here batch_size=320 i.e, considering the entire data
print(time.time() - start)
```

```python
# TIME TAKEN:
# Batch --> 0.8528134822845459
# Stochastic --> 10.792396545410156

# Batch faster
```

## 2. Accuracy test

```python
model = Sequential()

model.add(Dense(10, activation='relu', input_dim=2))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

#### Batch GD

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
# start = time.time()
history = model.fit(X_scaled, y, epochs=50, batch_size=400, validation_split=0.2) # here batch_size=320 i.e, considering the entire data
# print(time.time() - start)
```

#### Stochastic GD

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
# start = time.time()
history = model.fit(X_scaled, y, epochs=10, batch_size=1, validation_split=0.2) # here batch_size=320 i.e, considering the entire data
# print(time.time() - start)
```

**NOTE:**
1. Faster when model is runned for smae no of eposchs ---> Batch
2. Faster in converging i.e getting closer to the answer i.e better accuracy(in same epochs) --> Stochastic
