```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ushape.csv')
```

```python
df.head()
```

```python
plt.scatter(df['0'], df['1'], c=df['2'])
```

```python
X = df.iloc[:, :2].values
y = df['2'].values
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(10, activation='tanh', input_dim=2))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(10, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# to get all weights
model.get_weights()
```

```python
w_ls = list(model.get_weights())
for w in w_ls:
    print(w)
    print(len(w))
    print("--------------------------")
```

```python
initial_weights = model.get_weights()
```

```python
# initial_weights[0] = np.random.randn(2, 10)*np.sqrt(1/2) 
# initial_weights[1] = np.zeros(model.get_weights()[1].shape)
```

# Uniform 

```python
# now I am changing the weights

for i in range(len(initial_weights)):
    
    if(i%2 == 0):
        print(f"Layer - {i/2}:")
        # a. weights --> sqrt(1 / fan_in) | [fan_in --> no of incoming weights]
        w0, w1 = initial_weights[i].shape
        print(w0, w1)
        initial_weights[i] = np.random.randn(w0, w1)*np.sqrt(1/w0) # fan_in = w0
        print(f"Weights - {initial_weights[i]}")
    else:
        # b. bais --> initialize to zero
        initial_weights[i] = np.zeros(initial_weights[i].shape)
        print(f"Bias - {initial_weights[i]}")
        print("------------------------------------------------------------------------------------")
        
    
```

```python
# Setting the new weight values in the model
model.set_weights(initial_weights)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
history = model.fit(X, y, epochs=100, validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
